## 문제 1: 비밀번호 강도 체크
사용자로부터 비밀번호를 입력받아, 그 비밀번호가 강한지 약한지를 평가하는 프로그램을 작성하세요. 강한 비밀번호의 조건은 다음과 같습니다.

> 조건1. 8자리 이상

> 조건2. 숫자와 영문자를 반드시 포함해야 함. (대소문자는 상관X)

위의 조건을 만족하지 않을시, 약한 비밀번호로 지정합니다.

[HINT] str.isalpha(), str.isdigit(), str.isalnum()을 사용할 수 있습니다.

```python
# A1.
pw = 'abcd1234'
# if len(pw) < 8:
#     print('비밀번호는 8자리 이상이어야 합니다.')

# pw.isalpha()  # False


if pw.isalnum() and len(pw) >= 8:
    print(f'{pw} 강한 비밀번호입니다.')
else:
    print(f'{pw} 약한 비밀번호입니다.')


result = '강한 비밀번호입니다.' if pw.isalnum() and len(pw) >= 8 else '약한 비밀번호입니다.'
print(f'{pw} {result}')











```

## 문제 3. 단어 갯수 구하기

* 다음과 같은 영어 문서 하나가 있다.
* 다음 문서에 등장하는 단어마다 등장 횟수를 세어보자!

```python
pep8 = """This document gives coding conventions for the Python code comprising the standard library in the main Python distribution. Please see the companion informational PEP describing style guidelines for the C code in the C implementation of Python.
This document and PEP 257 (Docstring Conventions) were adapted from Guido’s original Python Style Guide essay, with some additions from Barry’s style guide [2].
This style guide evolves over time as additional conventions are identified and past conventions are rendered obsolete by changes in the language itself.
Many projects have their own coding style guidelines. In the event of any conflicts, such project-specific guides take precedence for that project.
"""

# 각 단어별로 몇 번씩 등장하는지 모두 출력하면 됩니다. 순서는 크게 상관없습니다. 텍스트 레벨에서의 전처리는 하지 않아도 됩니다.


# python dictionary indexing
# str.split()  # 공백을 기준으로 문자열을 나눕니다.
# for statement, list comprehension

pep8_words = pep8.split()  # 문자열을 공백 기준으로 나눕니다.
word_count = {}
for word in pep8_words:
    word = word.lower()  # 대소문자 구분을 없애기 위해 소문자로 변환
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# for word, count in word_count.items():
#     print(f"{word}: {count}")

sorted_word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_word_count:
    print(f"{word}: {count}")

```

```python
pep8 = """This document gives coding conventions for the Python code comprising the standard library in the main Python distribution. Please see the companion informational PEP describing style guidelines for the C code in the C implementation of Python.
This document and PEP 257 (Docstring Conventions) were adapted from Guido’s original Python Style Guide essay, with some additions from Barry’s style guide [2].
This style guide evolves over time as additional conventions are identified and past conventions are rendered obsolete by changes in the language itself.
Many projects have their own coding style guidelines. In the event of any conflicts, such project-specific guides take precedence for that project.
"""

comprehension = {word.lower(): pep8.split().count(word) for word in pep8.split()}
sorted_comprehension = sorted(comprehension.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_comprehension:
    print(f"{word}: {count}")   


```
